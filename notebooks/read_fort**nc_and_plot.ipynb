# Read Fort.**.nc and plot global outputs

### input paths to the fort files to generate 2d plots

```python
import matplotlib as mpl
mpl.use('Agg')
import matplotlib.pyplot as plt # Import the Matplotlib package
import numpy.core.multiarray 
from mpl_toolkits.basemap import Basemap
import numpy as np
import warnings
import matplotlib.cbook
warnings.filterwarnings("ignore",category=matplotlib.cbook.mplDeprecation)
from datetime import datetime, timedelta
import netCDF4
import matplotlib.tri as tri
from importlib import reload
import utils; reload(utils)
from utils import *
```

```python
fort22 = r'/scratch/04952/tmiesse/Dewberry/north_carolina/gustav/fort.221'
start_date = start_date(fort22)
```

```python
fort63 = r'/scratch/04952/tmiesse/Dewberry/north_carolina/gustav/fort.63.nc'
file,lat,lon,gridvars = fort_nc(fort63)
```

```python

#Longitudinal Correction
lon[lon < 180] = lon[lon < 180] + 360
var_element = 'element'
elems = gridvars[var_element][:,:]-1  # Move to 0-indexing by subtracting 1, elements indexing starts with '1' in netcdf file

# basemap 
m = Basemap(projection='cyl',llcrnrlat=30,urcrnrlat=40,llcrnrlon=280,urcrnrlon=288,resolution='h', epsg = 4269)
#US is 4269, and you can google the region you want
```

```python
for i in range(0,90):
   
    i+=1 
    data = file.variables['zeta'][i,:]
    z = data.data
    file_number = '%02d'%i

    print ('\nTriangulating ...\n')
    triang = tri.Triangulation(lon,lat, triangles=elems)

    if data.mask.any():
      # -99999 entries in 'data' array are usually masked, mask all corresponding triangles
      point_mask_indices = numpy.where(data.mask)
      tri_mask = numpy.any(numpy.in1d(elems, point_mask_indices).reshape(-1, 3), axis=1)
      triang.set_mask(tri_mask)

    # First create the x and y coordinates of the points.
  
    levels = np.arange(-1.5, 5, 0.1)
  
    #plt.scatter(lon, lat, s=0.0125, c=z,cmap='jet')
        
    print ('Making contours ...\n')
    plt.tricontourf(triang, data, levels=levels,alpha=0.9,vmin=-1.5, vmax=5, aspect='auto',cmap='jet')
 
    m.arcgisimage(service='World_Street_Map', xpixels = 800, verbose= False)
    #World_Street_Map,ESRI_Imagery_World_2D,NatGeo_World_Map ,World_Imagery (type of Data to use from)  
     
    #m.drawcoastlines(color='k')

    # Scatte point for capital and specified cities
    plt.title('Water Levels\n')
    #plt.xlim([282, 286])
    #plt.ylim([36, 40])    
    plt.colorbar(cmap='jet',label='Water Levels (meters realtive to MSL)',format = "%.1f")    
    #plt.xlabel('\n\nDate (UTC):{} \nForecast Hour: {} \n'.format(start_date+ timedelta(hours=i),file_number))   
    plt.savefig('/scratch/04952/tmiesse/Dewberry/north_carolina/gustav/outputs/WL{}.png'.format(file_number),
                dpi=300, bbox_inches = 'tight', pad_inches = 0.1)
    plt.close() 
```

## Plot the max water level

```python
max63 = r'/scratch/04952/tmiesse/Dewberry/north_carolina/surf_city/maxele.63.nc'
file,lat,lon,gridvars = fort_nc(max63)
```

```python

#Longitudinal Correction
lon[lon < 180] = lon[lon < 180] + 360
var_element = 'element'
elems = gridvars[var_element][:,:]-1  # Move to 0-indexing by subtracting 1, elements indexing starts 
                                      # with '1' in netcdf file

# basemap 
m = Basemap(projection='cyl',llcrnrlat=34.459,urcrnrlat=35.1,llcrnrlon=282.95,urcrnrlon=283.8,
            resolution='h', epsg = 4269)
#US is 4269, and you can google the region you want
```

```python
data = file.variables['zeta_max'][:]
z = data.data
file_number = '%02d'%1

print ('\nTriangulating ...\n')
triang = tri.Triangulation(lon,lat, triangles=elems)

if data.mask.any():
  # -99999 entries in 'data' array are usually masked, mask all corresponding triangles
  point_mask_indices = numpy.where(data.mask)
  tri_mask = numpy.any(numpy.in1d(elems, point_mask_indices).reshape(-1, 3), axis=1)
  triang.set_mask(tri_mask)

# First create the x and y coordinates of the points.

levels = np.arange(-1.5, 5, 0.1)

#plt.scatter(lon, lat, s=0.0125, c=z,cmap='jet')

print ('Making contours ...\n')
plt.tricontourf(triang, data, levels=levels,alpha=0.9,vmin=-1.5, vmax=5, aspect='auto',cmap='jet')

m.arcgisimage(service='World_Street_Map', xpixels = 800, verbose= False)
#World_Street_Map,ESRI_Imagery_World_2D,NatGeo_World_Map ,World_Imagery (type of Data to use from)  

#m.drawcoastlines(color='k')

# Scatte point for capital and specified cities
plt.title('Water Levels\n')
plt.xlim([282.95, 283.8])
plt.ylim([34.459, 35.1])    
plt.colorbar(cmap='jet',label='Water Levels (meters realtive to MSL)',format = "%.1f")    
#plt.xlabel('\n\nDate (UTC):{} \nForecast Hour: {} \n'.format(start_date+ timedelta(hours=i),file_number))   
plt.savefig('/scratch/04952/tmiesse/Dewberry/north_carolina/surf_city/output/max_WL{}.png'.format(file_number),
            dpi=300, bbox_inches = 'tight', pad_inches = 0.1)
plt.close()
```
