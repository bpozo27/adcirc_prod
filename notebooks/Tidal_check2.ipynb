### Beaufort

```python
nc_file = nc4.Dataset(f61)
start = '20171213 12:20:00'
freq = '6min'
model_data = nc_file['zeta'][:,0]
model1 = pd.DataFrame(model_data)
date   = pd.date_range(start=start,periods=int(len(model1)),freq=freq)
model1.insert(0,'Date Time',date)
model1 = model1.rename(columns={0:' Water Level'})
obs = pd.read_csv(os.path.join(root_dir,'Duke_20171213_20180113_obs_msl.csv'))
#obs2 = pd.read_csv(os.path.join(root_dir,'Duke_20171213_20180113_navd.csv'))
obs3 = pd.read_csv(os.path.join(root_dir,'Duke_20180110_20180201_obs_msl.csv'))
pred = pd.read_csv(os.path.join(root_dir,'Duke_20171213_20180113.csv'))
pred2 = pd.read_csv(os.path.join(root_dir,'Duke_20180110_20180201.csv'))
```

```python
trace1 = go.Scatter(
        x = model1['Date Time'],
        y = model1[' Water Level']-.1,
        name = 'ADCIRC Beaufort',
        mode = 'lines',
        line=dict(color= 'rgb(0, 204, 255)'))
trace2 = go.Scatter(
        x = obs['Date Time'],
        y = obs[' Water Level'],
        name = 'Observed Beaufort',
        mode = 'lines',
        line=dict(color= 'rgb(51, 51, 255)'))
trace3 = go.Scatter(
        x = obs3['Date Time'],
        y = obs3[' Water Level'],
        name = 'Observed Beaufort',
        mode = 'lines',
        line= dict(color= 'rgb(51, 51, 255)'))
trace4 = go.Scatter(
        x = pred['Date Time'],
        y = pred[' Prediction'],
        name = 'Pred Beaufort',
        mode = 'lines',
        line=dict(color= 'rgb(151, 51, 255)'))
trace5 = go.Scatter(
        x = pred2['Date Time'],
        y = pred2[' Prediction'],
        name = 'Pred Beaufort',
        mode = 'lines',
        line= dict(color= 'rgb(151, 51, 255)'))
data = [trace1,trace4,trace5]
layout = go.Layout(dict(title = 'Water level'),
          xaxis = dict(title = 'Time'),
          yaxis = dict(title = 'Water level (m) at MTL'),                     
          legend= dict(orientation = "h"),
          font = dict(color = 'rgb(255,255,255)'),
          paper_bgcolor = 'rgb(0,0,0)',
          plot_bgcolor = 'rgb(0,0,0)')
fig = go.Figure(data=data, layout=layout)
interactive = iplot(fig,filename=root_dir+'\\'+'beaufort_20171213_20180201.html')
plot(fig,filename=root_dir+'\\'+'beaufort_20171213_20180201.html')
```
