```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
time = np.arange(0.0, 60, 0.25) # total time period of total 1 minute with a half second period
```

```python
fig, ax = plt.subplots(figsize=(18,6))
ax.plot(time, np.sin(2*np.pi*time),'k')
ax.set_xlim(0,60)
ax.set_title('Sin Wave Plot')
ax.set_xlabel('Time (s)')
ax.set_ylabel('Hz')
```
